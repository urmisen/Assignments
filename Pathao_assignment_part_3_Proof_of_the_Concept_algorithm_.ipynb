```python
dataset = pd.read_csv('/content/drive/My Drive/pathao_assignment/Driver_Profilling_Dataset.csv')

#print the dataset shape with rows and columns
print(dataset.shape)
```

```python
#print the top-5 records
dataset.head()
```

```python
driver_id= list(dataset['Driver_ID'])
Average_ride_distance = list(dataset['Average_Ride_Distance'])
ACR = list(dataset['Assignment_to_completion_ratio'])
Average_ride_duration = list(dataset['Average_Ride_Duration'])
Avg_Rating = list(dataset['Average_Rating'])
print(driver_id[0])
```

```python
Req_Dist = input("Enter the value of the requested riding distance :")
print(Req_Dist)
```

```python
index=[]
key=[]
diff=[]
# Finding the list of difference to the given Riding Distance to each of the Average Riding Distance for each of the Drivers.

for i in range(len(Average_ride_distance)):
  x= abs(int(Req_Dist) - (Average_ride_distance[i]))
  diff.append(x)
  key.append(i)

res = {key[i]: diff[i] for i in range(len(diff))}
print(res)

# Here Key keeps track of the indexes
# and diff keeps track of the differences


```

```python
#sorting the dictionary

sorted_values = sorted(res.values()) # Sort the values
sorted_dict = {}

for i in sorted_values:
    for k in res.keys():
        if res[k] == i:
            sorted_dict[k] = res[k]
            break

print(sorted_dict)

```

```python
sum(Avg_Rating)/len(Avg_Rating)
```

```python
# taking the sorted indexes
index= list(sorted_dict.keys())
print(index)

ls=[]
driv=[]

for i in range(0,10):   #  taking top 10 closed riding Distance
  x=index[1]
  ls.append(ACR[x])
  driv.append(driver_id[x])
ind=[i for i, j in enumerate(ls) if j == max(ls)]
print(ind[0])
if(Avg_Rating[ind[0]] > (sum(Avg_Rating)/len(Avg_Rating))):
  k=ind[0]
  
assigned_driver = driver_id[k]

```

```python
print("The list of the closest Drivers is :",driv)
print("\n")
print("Highest probability of accepting the request is with :",assigned_driver)
print("\n")
print("With Assignment to completion ratio (ACR) :", max(ls))
print("\n")
print("Where the Aveerage Rating of the driver given by user at the end of the trip :", Avg_Rating[ind[0]])
```
